### Example
_Creating the following plots_

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
textstr1 = '\n'.join((
            r'$\sigma_a= %.4f$' % (np.std(res_a1)),
            r'$\sigma_b= %.4f$' % (np.std(res_b1))))
textstr2 = '\n'.join((
            r'$\sigma_a= %.4f$' % (np.std(res_a2)),
            r'$\sigma_b= %.4f$' % (np.std(res_b2))))
```

```python
textstr1
```

Make plot

```python
fig, axes = plt.subplots(1,2, figsize = (10,3.5))
ax = axes[0]
ax.set_title("Plot 1")
ax.hist(res_a1, histtype = 'step', density = True, label = 'a', bins = 30, color = 'blue')
ax.hist(res_b1, histtype = 'step', density = True, label = 'b', bins = 30,color = 'red')
ax.legend(loc = 'upper right', fancybox = False, edgecolor ='black')
ax.text(0.05, 0.81, textstr1, transform = ax.transAxes, bbox = dict(facecolor = 'white', edgecolor = 'black'))
ax.set_ylabel("Density")

ax = axes[1]
ax.set_title("Plot 2")
ax.hist(res_a2, histtype = 'step', density = True, label = 'a', bins = 30, color = 'green')
ax.hist(res_b2, histtype = 'step', density = True, label = 'b', bins = 30, color = 'purple')
ax.legend(loc = 'upper right', fancybox = False, edgecolor ='black')
ax.text(0.05, 0.81, textstr2, transform = ax.transAxes, bbox = dict(facecolor = 'white', edgecolor = 'black'))
fig.text(0.5 , -0.04, '$\Delta E$ [Joules]', ha = 'center', fontsize = 14)
plt.show()
```

## Stream Plot

For vector plot

```python
w= 3
_ = np.linspace(-3,3, 100)
X, Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2,figsize = (7,7))
ax = axes[0][0]
ax.streamplot(X,Y,U,V)
             
ax = axes[0][1]             
ax.streamplot(X,Y,U,V, color = speed)     # imagine the plot is the flow of water

ax = axes[1][0]  
lw = 5*speed/speed.max()
ax.streamplot(X,Y,U,V, linewidth = lw)

ax = axes[1][1]  
lw = 5*speed/speed.max()
seedpoints = np.array([[0,1],[1,0]])
ax.streamplot(X,Y,U,V, start_points = seedpoints)
```

### Animation using Matplotlib

Animation require relating the time of your problem to the frame of the animation, then specifying the frames per second
- Below is the sin wave moving at 3m/s. Translate this an animation

```python
def f(x,t):
    return np.sin(x-3*t)
x = np.linspace(0, 10*np.pi, 1000)
```

Create Animation

```python
from matplotlib import animation
from matplotlib.animation import PillowWriter
```

```python
fig, ax = plt.subplots(1,1, figsize = (8,4))
ln1, = plt.plot([],[])
time_text = ax.text(0.65, 0.95,'', fontsize = 50,
                    transform = ax.transAxes, bbox = dict(facecolor = 'white', edgecolor = 'black'))
ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5,1.5)
def animate(i):
    ln1.set_data(x,f(x,1/50*i))
    time_text.set_text('t = {:.2f}'.format(i/50))

ani = animation.FuncAnimation(fig, animate, frames = 240, interval = 50)
ani.save('ani.gif',writer = 'pillow',fps = 50, dpi = 100)
```
